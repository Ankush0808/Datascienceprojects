```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np                                                                                                                                                              
```

```python
df=pd.read_csv("C://Users//Lenovo//Documents//deliveries.csv")
```

```python
df[
    (df['batting_team'].isin(["Royal Challengers Bangalore", "Royal Challengers Bengaluru"])) |
    (df['bowling_team'].isin(["Royal Challengers Bangalore", "Royal Challengers Bengaluru"]))
]
```

```python
df_m=pd.read_csv("C://Users//Lenovo//Documents//matches.csv")
```

```python
df_m
```

```python
df_m[
    (df_m['team1'].isin(["Royal Challengers Bangalore", "Royal Challengers Bengaluru"])) |
    (df_m['team2'].isin(["Royal Challengers Bangalore", "Royal Challengers Bengaluru"]))
]
```

```python
df_final=pd.merge(df,df_m,how="inner",left_on="match_id",right_on="id")
```

```python
df_final
```

```python
df_final['current_score'] = df_final.groupby(['match_id','inning'])['total_runs'].cumsum()
```

```python
df_final['Current_wickets']=df_final.groupby(['match_id','inning'])['is_wicket'].cumsum()
```

```python
df_final_chase=df_final[df_final['inning']==2]
```

```python
df_model1=df_final_chase[["batting_team","bowling_team","over","ball","Current_wickets","current_score","target_runs","target_overs","venue","winner"]]
```

```python
df_model1
```

```python
df_model1['winner']=np.where(df_model1['winner']==df_model1['batting_team'],1,0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
X=df_model1.drop(columns=['winner'])
y=df_model1['winner']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler,OrdinalEncoder,LabelEncoder
```

```python
trf1=ColumnTransformer([('scaling_numerical',StandardScaler(),[2,3,4,5,6,7])],remainder='passthrough')
```

```python

trf2 = ColumnTransformer(
    transformers=[('ohe_trans', OneHotEncoder(sparse_output=False),[0,1,8])],
    remainder='passthrough'
)
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
trf3=SelectKBest(score_func=chi2,k=20)
```

```python
trf4=LogisticRegression(class_weight='balanced')
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
pipe=make_pipeline(trf2,trf1,trf4)
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.named_steps
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import  accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(pipe,X_train,y_train,cv=5,scoring='accuracy').mean()
```

```python
from sklearn.metrics import confusion_matrix,precision_score,f1_score,recall_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
recall_score(y_test,y_pred)
```
