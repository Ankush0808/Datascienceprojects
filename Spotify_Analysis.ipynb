# Feature transformation

```python
df_capping.skew()
```

```python
df_capping[list1[8]]=np.where(df_capping[list1[8]]==0,df_capping[list1[8]]+0.001,df_capping[list1[8]])
```

```python
# Now therefore the attributes with being left skewed are transformed using log transformtion and for 
#right skewed the square or power transformation is used
```

```python
from sklearn.preprocessing import FunctionTransformer,PowerTransformer
```

```python
ft=FunctionTransformer(np.log1p)
pt=PowerTransformer()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer(transformers=[
    ('ft',FunctionTransformer(np.log1p),[2,8]),
    ('pt',PowerTransformer(),[0,1,3,4,5,6,7])
],remainder='passthrough')
```

```python
df_capping=ct.fit_transform(df_capping)
```

```python
df_capping
```

```python
from sklearn.cluster import KMeans
```

```python

k=[]
iner=[]
for i in range(1,25):
    kmeans=KMeans(n_clusters=i).fit(df_capping)
    iner.append(kmeans.inertia_)
    k.append(i)
plt.plot(k,iner)
    
```

```python
kmeans=KMeans(n_clusters=4)
y_capp_trial=kmeans.fit_predict(df_capping)
```

```python
from sklearn.metrics import silhouette_score
```

```python
ss=silhouette_score(df_capping,y_capp_trial)
```

```python
ss
```

```python
param_grid={
    'init':['k-means++','random'],
    'n_clusters':[2,3,4,5,6,7,8],
    'max_iter':[50,100,150,200,300,350,400]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grd=GridSearchCV(estimator=kmeans,
                param_grid=param_grid,
                cv=5)
```

```python
grd.fit(df_capping)
```

```python
grd.best_params_
```

```python
kmeans2=KMeans(n_clusters=4,max_iter=150,init='k-means++')
y_capp2_trial=kmeans2.fit_predict(df_capping)
```
